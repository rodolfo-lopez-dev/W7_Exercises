```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots


```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Styling Plots


```python
plt.plot(x, y, linestyle='--', color='red')
plt.show()
```

```python
plt.plot(x, y, marker='s', ms=10, mfc='green')
plt.show()
```

```python
plt.plot(x, y, marker='D', ms=14, mfc='purple')
plt.plot(x, y, marker='d', ms=10, mfc='white')
plt.show()
```

## Advanced Plotting Techniques


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]
```

```python
plt.bar(products, sales, color='skyblue', width=0.5)
plt.title('Sales by Product')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, s=np.random.rand(50) * 100, c=np.random.rand(50), cmap='viridis')
plt.colorbar()
plt.title('Scatter Plot Example')
plt.xlabel('Random X')
plt.ylabel('Random Y')
plt.show()
```

## Subplots


```python
# Sine Wave
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Plot for above
plt.plot(x, y, color='blue', linestyle='-')
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.show()
```

```python
# Data for Histogram
data = np.random.randn(1000)
```

```python
# Plot the Histogram
plt.hist(data, bins=20, color='white', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
